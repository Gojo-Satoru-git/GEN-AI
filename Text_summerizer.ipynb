```python
# Install necessary libraries
# pip install gradio transformers

import gradio as gr
from transformers import pipeline

# Create a summarization pipeline using a pre-trained model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Define the function for text summarization
def summarize_text(input_text):
    # Generate the summary with specified parameters
    summary = summarizer(input_text, max_length=130, min_length=30, do_sample=False)
    return summary[0]["summary_text"]

# Create a Gradio interface
interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize..."),
    outputs=gr.Textbox(lines=5, label="Summary"),
    title="Text Summarization App",
    description="This app summarizes long pieces of text using a pre-trained model from Hugging Face."
)

# Launch the Gradio app
interface.launch()
```
