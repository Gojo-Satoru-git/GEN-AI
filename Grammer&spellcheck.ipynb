```python
import gradio as gr
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the model and tokenizer from Hugging Face
model_name = "samadpls/t5-base-grammar-checker"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Define the function for grammar correction
def grammar_spell_check(input_text):
    # Prepare the input for the model
    input_text = "grammar: " + input_text
    inputs = tokenizer.encode(input_text, return_tensors="pt")

    # Generate corrected output
    outputs = model.generate(inputs)
    corrected_sentence = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return corrected_sentence

# Create a Gradio interface
interface = gr.Interface(
    fn=grammar_spell_check,
    inputs=gr.Textbox(lines=5, placeholder="Enter text with grammatical errors here..."),
    outputs="text",
    title="Grammar and Spell Checker",
    description="Enter a sentence or paragraph, and this tool will correct grammatical errors and spelling mistakes."
)

# Launch the Gradio app
interface.launch()
```
