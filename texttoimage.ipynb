```python
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt



headers = {"Authorization": f"Bearer {API_TOKEN}"}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    print("Response Status Code:", response.status_code)  # Debugging

    if response.status_code != 200:
        print("Error Response:", response.text)  # Print error if any
        raise Exception(f"Request failed with status {response.status_code}")

    return response.content

# Define your text prompt
prompt = "A cat flying over neon city"
payload = {"inputs": prompt}

# Query the API to generate the image
image_data = query(payload)

# Check if API returned valid image data
if len(image_data) < 1000:  # If data size is too small, it's probably an error
    print("API returned invalid image data!")
else:
    # Convert binary data to an image
    image = Image.open(BytesIO(image_data))

    # Save the image to a file
    image.save("output.png")
    print("Image saved as output.png")

    # Display using Matplotlib
    plt.imshow(image)
    plt.axis("off")
    plt.show()

```
